```python
#Compare reference audio and the one generated by the network + synth
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_power_spectrogram(signal, sr, title):
    power_spec = librosa.power_to_db(librosa.feature.melspectrogram(y=signal, sr=sr), ref=np.max)
    librosa.display.specshow(power_spec, sr=sr, x_axis='time', y_axis='mel')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.tight_layout()


```

```python
ref = 'comparisons/bass_electronic_003-055-127.wav'
synth = 'comparisons/tensor_audio.wav'


ref, sr1 = librosa.load(ref)
synth, sr2 = librosa.load(synth)

plt.figure(figsize=(14, 4))

plt.subplot(1, 2, 1)
plot_power_spectrogram(ref, sr1, 'Power Spectrogram - Reference signal')

plt.subplot(1, 2, 2)
plot_power_spectrogram(synth, sr2, 'Power Spectrogram - Synthesised signal')

plt.savefig('comparison_1.svg')
plt.show()

```
